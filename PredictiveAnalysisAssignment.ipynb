## Step 2: Data Cleaning

```python
#Deleting the unneccesary column "car_ID"
car_pricing.drop(['car_ID'], axis = 1, inplace = True)
```

## Data Preparation

```python
#Splitting the CarName for the company name only.
car_pricing['CarName'] = car_pricing.CarName.str.split(' ', expand = True)
car_pricing.head(5)
```

```python
# Defining the map function to convert binary categorical variables
def binary_map_doornumber(x):
    return 0 if x == "two" else 1

def binary_map_fueltype(x):
    return 0 if x == "gas" else 1

def binary_map_aspiration(x):
    return 0 if x == "std" else 1

def binary_map_enginelocation(x):
    return 0 if x == "front" else 1

def binary_map_cylinderNumber(x):
    return 2 if x == "two" else (3 if x == "three" else (4 if x == "four" else (5 if x == "five" else (6 if x == "six" else (8 if x == "eight" else (12))))))


car_pricing['doornumber'] = car_pricing['doornumber'].apply(binary_map_doornumber)
car_pricing['fueltype'] = car_pricing['fueltype'].apply(binary_map_fueltype)
car_pricing['aspiration'] = car_pricing['aspiration'].apply(binary_map_aspiration)
car_pricing['enginelocation'] = car_pricing['enginelocation'].apply(binary_map_enginelocation)
car_pricing['cylindernumber'] = car_pricing['cylindernumber'].apply(binary_map_cylinderNumber)
```

### Creating Dummy variables

```python
# Get the dummy variables for the feature 'carbody' and store it in a new variable - 'carbody'
carBody = pd.get_dummies(car_pricing['carbody'], drop_first = True)
```

```python
# Get the dummy variables for the feature 'drivewheel' and store it in a new variable - 'driveWheel'
driveWheel = pd.get_dummies(car_pricing['drivewheel'], drop_first = True)
```

```python
# Get the dummy variables for the feature 'CarName' and store it in a new variable - 'CarName'
CarName = pd.get_dummies(car_pricing['CarName'], drop_first = True)
```

```python
# Get the dummy variables for the feature 'fuelsystem' and store it in a new variable - 'fuelSystem'
fuelSystem = pd.get_dummies(car_pricing['fuelsystem'], drop_first = True)
```

```python
# Get the dummy variables for the feature 'enginetype' and store it in a new variable - 'enginetype'
enginetype = pd.get_dummies(car_pricing['enginetype'], drop_first = True)
```

```python
# Add the results to the original housing dataframe
car_pricing = pd.concat([car_pricing, carBody, driveWheel, CarName, fuelSystem, enginetype], axis = 1)
```

```python
# Now let's see the head of our dataframe.

car_pricing.head()
```

```python
# Dropping all those columns for which we have created the dummies

car_pricing.drop(['carbody', 'drivewheel', 'CarName', 'fuelsystem', 'enginetype'], axis = 1, inplace = True)
```

```python
car_pricing.head()
```

## Splitting the Data into Training and Testing Sets

```python
from sklearn.model_selection import train_test_split

# We specify this so that the train and test data set always have the same rows, respectively
np.random.seed(0)
df_train, df_test = train_test_split(car_pricing, train_size = 0.7, test_size = 0.3, random_state = 100)
```

### Rescaling the Features
We will use MinMax scaling.

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

# Applyin scaler() to all the columns except the binary categorical and 'dummy' variables
num_vars = ['symboling', 'wheelbase', 'carlength', 'cylindernumber', 'carwidth', 'carheight', 'curbweight', 'compressionratio', 'horsepower', 'peakrpm', 'citympg', 'highwaympg', 'price', 'enginesize', 'boreratio', 'stroke']

df_train[num_vars] = scaler.fit_transform(df_train[num_vars])
```
